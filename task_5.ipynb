## Problem 1: Social Media – Post Engagement Analyzer

**Problem Statement:** Create a function to analyze engagement on a social media post.

**Rules:**
*   Calculate total likes
*   If Total Likes ≥ 1000 → Viral Post
*   Otherwise → Normal Engagement

**Expected Output Format:**
```
Total Likes: 1050
Post Status: Viral Post
```

````python
def analyze_post_engagement(likes):
    """
    Analyzes social media post engagement based on the number of likes.

    Args:
        likes (int): The total number of likes on the post.

    Returns:
        tuple: A tuple containing the total likes and the post status.
    """
    total_likes = likes
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"
    return total_likes, post_status

# Example Usage:
likes_1 = 1050
total_likes_1, status_1 = analyze_post_engagement(likes_1)
print(f"Total Likes: {total_likes_1}")
print(f"Post Status: {status_1}")

print("\n---")

likes_2 = 500
total_likes_2, status_2 = analyze_post_engagement(likes_2)
print(f"Total Likes: {total_likes_2}")
print(f"Post Status: {status_2}")
````

## Problem 3: Agriculture – Rainfall Adequacy Checker

**Problem Statement:** Create a function that analyzes seasonal rainfall data.

**Rules:**
*   Calculate average rainfall
*   Average ≥ required level → Adequate Rainfall
*   Otherwise → Inadequate Rainfall

**Expected Output Format:**
```
Average Rainfall: 72
Rainfall Status: Adequate Rainfall
```

````python
def analyze_rainfall(rainfall_data, required_level):
    """
    Analyzes seasonal rainfall data to determine adequacy.

    Args:
        rainfall_data (list): A list of rainfall measurements (e.g., in mm).
        required_level (int/float): The minimum required average rainfall level.

    Returns:
        tuple: A tuple containing the average rainfall and its status.
    """
    if not rainfall_data:
        return 0, "No data to analyze"

    total_rainfall = 0
    for rain in rainfall_data:
        total_rainfall += rain
    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return round(average_rainfall, 2), status

# Example Usage:
rainfall_1 = [60, 75, 80, 70, 75]
required_level_1 = 70
avg_rainfall_1, status_1 = analyze_rainfall(rainfall_1, required_level_1)
print(f"Average Rainfall: {avg_rainfall_1}")
print(f"Rainfall Status: {status_1}")

print("\n---")

rainfall_2 = [50, 60, 45, 55]
required_level_2 = 70
avg_rainfall_2, status_2 = analyze_rainfall(rainfall_2, required_level_2)
print(f"Average Rainfall: {avg_rainfall_2}")
print(f"Rainfall Status: {status_2}")
````

## Problem 4: Social Media – Duplicate Account Detection

**Problem Statement:** Create a function to detect duplicate usernames in a system.

**Rules:**
*   If duplicates exist → Flag accounts
*   Else → Accounts are unique

**Expected Output Format:**
```
Duplicate Accounts Found: Yes
```

````python
def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames in a list of accounts.

    Args:
        usernames (list): A list of usernames.

    Returns:
        str: A string indicating whether duplicate accounts were found.
    """
    unique_usernames = set(usernames)
    if len(usernames) != len(unique_usernames):
        return "Duplicate Accounts Found: Yes"
    else:
        return "Accounts are unique"

# Example Usage:
user_list_1 = ["alice", "bob", "charlie", "alice"]
duplicate_status_1 = detect_duplicate_accounts(user_list_1)
print(duplicate_status_1)

print("\n---")

user_list_2 = ["david", "eve", "frank"]
duplicate_status_2 = detect_duplicate_accounts(user_list_2)
print(duplicate_status_2)
````

## Problem 6: Agriculture – Premium Crop Price Filter

**Problem Statement:** Create a function that filters crop prices above a premium threshold.

**Rules:**
*   Price > ₹2000 → Premium Crop
*   Else → Regular Crop

**Expected Output Format:**
```
Premium Crops: [2500, 3200]
```

````python
def filter_premium_crops(prices):
    """
    Filters crop prices above a premium threshold.

    Args:
        prices (list): A list of crop prices.

    Returns:
        list: A list of premium crop prices.
    """
    premium_crops = [price for price in prices if price > 2000]
    return premium_crops

# Example Usage:
crop_prices_1 = [1800, 2500, 1500, 3200, 2000]
premium_crops_1 = filter_premium_crops(crop_prices_1)
print(f"Premium Crops: {premium_crops_1}")

print("\n---")

crop_prices_2 = [1000, 1500, 1900]
premium_crops_2 = filter_premium_crops(crop_prices_2)
print(f"Premium Crops: {premium_crops_2}")
````

## Problem 7: System Monitoring – Application Health Checker

**Problem Statement:** Create a function to evaluate application health based on error count.

**Rules:**
*   Errors = 0 → Healthy
*   Errors ≤ 5 → Minor Issues
*   Errors > 5 → Critical Issues

**Expected Output Format:**
```
Error Count: 7
System Status: Critical Issues
```

````python
def check_application_health(error_count):
    """
    Evaluates application health based on the number of errors.

    Args:
        error_count (int): The number of errors detected.

    Returns:
        tuple: A tuple containing the error count and system status.
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return error_count, status

# Example Usage:
errors_1 = 7
error_count_1, status_1 = check_application_health(errors_1)
print(f"Error Count: {error_count_1}")
print(f"System Status: {status_1}")

print("\n---")

errors_2 = 3
error_count_2, status_2 = check_application_health(errors_2)
print(f"Error Count: {error_count_2}")
print(f"System Status: {status_2}")

print("\n---")

errors_3 = 0
error_count_3, status_3 = check_application_health(errors_3)
print(f"Error Count: {error_count_3}")
print(f"System Status: {status_3}")
````

## Problem 8: Banking – Daily Transaction Limit Checker

**Problem Statement:** Create a function to validate daily transaction amount.

**Rules:**
*   Daily limit = ₹50,000
*   Amount within limit → Approved
*   Amount exceeds limit → Rejected

**Expected Output Format:**
```
Transaction Amount: 60000
Transaction Status: Rejected
```

````python
def check_transaction_limit(amount, daily_limit=50000):
    """
    Validates a daily transaction amount against a set limit.

    Args:
        amount (float): The transaction amount.
        daily_limit (float): The maximum allowed daily transaction limit.

    Returns:
        tuple: A tuple containing the transaction amount and its status.
    """
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    return amount, status

# Example Usage:
transaction_1 = 60000
amount_1, status_1 = check_transaction_limit(transaction_1)
print(f"Transaction Amount: {amount_1}")
print(f"Transaction Status: {status_1}")

print("\n---")

transaction_2 = 45000
amount_2, status_2 = check_transaction_limit(transaction_2)
print(f"Transaction Amount: {amount_2}")
print(f"Transaction Status: {status_2}")
````

## Problem 9: E-Learning – Student Attendance Eligibility System

**Problem Statement:** Create a function to calculate attendance percentage and check exam eligibility.

**Rules:**
*   Attendance ≥ 75% → Eligible
*   Attendance < 75% → Not Eligible

**Expected Output Format:**
```
Attendance Percentage: 80.0
Exam Eligibility: Eligible
```

````python
def check_exam_eligibility(total_classes, classes_attended):
    """
    Calculates attendance percentage and checks exam eligibility.

    Args:
        total_classes (int): The total number of classes conducted.
        classes_attended (int): The number of classes attended by the student.

    Returns:
        tuple: A tuple containing the attendance percentage and exam eligibility status.
    """
    if total_classes == 0:
        return 0.0, "Not Applicable (No classes held)"

    attendance_percentage = (classes_attended / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    return round(attendance_percentage, 1), eligibility

# Example Usage:
total_classes_1 = 100
classes_attended_1 = 80
percentage_1, eligibility_1 = check_exam_eligibility(total_classes_1, classes_attended_1)
print(f"Attendance Percentage: {percentage_1}")
print(f"Exam Eligibility: {eligibility_1}")

print("\n---")

total_classes_2 = 120
classes_attended_2 = 85
percentage_2, eligibility_2 = check_exam_eligibility(total_classes_2, classes_attended_2)
print(f"Attendance Percentage: {percentage_2}")
print(f"Exam Eligibility: {eligibility_2}")
````

## Problem 10: Smart Electricity Bill Analyzer

**Problem Statement:** Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules:**
*   First 100 units → ₹3 per unit
*   Next 100 units (101–200) → ₹5 per unit
*   Above 200 units → ₹7 per unit

**Usage Classification:**
*   Total bill < ₹500 → Low Usage
*   ₹500 – ₹1500 → Moderate Usage
*   ₹1500 – ₹2500 → High Usage
*   > ₹2500 → Very High Usage


````python
def analyze_electricity_bill(units_consumed):
    """
    Calculates the electricity bill and classifies usage level.

    Args:
        units_consumed (int): The number of electricity units consumed.

    Returns:
        tuple: A tuple containing the total bill and the usage classification.
    """
    total_bill = 0
    remaining_units = units_consumed

    # Calculate bill based on tiers
    if remaining_units > 0:
        if remaining_units <= 100:
            total_bill += remaining_units * 3
            remaining_units = 0
        else:
            total_bill += 100 * 3
            remaining_units -= 100

    if remaining_units > 0:
        if remaining_units <= 100: # Units from 101 to 200
            total_bill += remaining_units * 5
            remaining_units = 0
        else:
            total_bill += 100 * 5
            remaining_units -= 100

    if remaining_units > 0: # Units above 200
        total_bill += remaining_units * 7

    # Classify usage
    if total_bill < 500:
        usage_classification = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_classification = "Moderate Usage"
    elif 1500 < total_bill <= 2500:
        usage_classification = "High Usage"
    else:
        usage_classification = "Very High Usage"

    return total_bill, usage_classification

# Example Usage:
units_1 = 80
bill_1, usage_1 = analyze_electricity_bill(units_1)
print(f"Units Consumed: {units_1}")
print(f"Total Bill: ₹{bill_1}")
print(f"Usage Classification: {usage_1}")

print("\n---")

units_2 = 150
bill_2, usage_2 = analyze_electricity_bill(units_2)
print(f"Units Consumed: {units_2}")
print(f"Total Bill: ₹{bill_2}")
print(f"Usage Classification: {usage_2}")

print("\n---")

units_3 = 250
bill_3, usage_3 = analyze_electricity_bill(units_3)
print(f"Units Consumed: {units_3}")
print(f"Total Bill: ₹{bill_3}")
print(f"Usage Classification: {usage_3}")

print("\n---")

units_4 = 400
bill_4, usage_4 = analyze_electricity_bill(units_4)
print(f"Units Consumed: {units_4}")
print(f"Total Bill: ₹{bill_4}")
print(f"Usage Classification: {usage_4}")
````
