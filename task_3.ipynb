### Problem Statement 1: Employee Performance Bonus Eligibility

This code identifies employees with the highest performance score from a given dictionary. It handles cases where multiple employees have the same highest score and displays all eligible top performers.

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
highest_score = 0
for score in employees.values():
    if score > highest_score:
        highest_score = score

# Identify all employees with the highest score
top_performers = []
for employee, score in employees.items():
    if score == highest_score:
        top_performers.append(employee)

# Format the output
output_string = "Top Performers Eligible for Bonus: "
output_string += ", ".join(top_performers)
output_string += f" (Score: {highest_score})"

print(output_string)
```

### Problem Statement 2: Search Query Keyword Analysis

This code analyzes a customer search query. It converts the query to lowercase, removes common punctuation, and then counts the frequency of each keyword. Finally, it displays only those keywords that appeared more than once.

```python
import string

search_query = "Buy mobile phone buy phone online"

# Convert to lowercase
search_query = search_query.lower()

# Remove punctuation
# Create a translation table to replace punctuation with spaces
translation_table = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
search_query = search_query.translate(translation_table)

# Split into words and count frequency
words = search_query.split()
keyword_counts = {}
for word in words:
    keyword_counts[word] = keyword_counts.get(word, 0) + 1

# Display keywords searched more than once
result = {word: count for word, count in keyword_counts.items() if count > 1}
print(result)
```

### Problem Statement 3: Sensor Data Validation

This code processes a list of sensor readings. It iterates through the readings, identifies those that are even numbers (considered valid), and stores them as `(hour_index, reading_value)` pairs. Odd readings are ignored.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []
for i, reading in enumerate(sensor_readings):
    if reading % 2 == 0:
        valid_readings.append((i, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

### Problem Statement 4: Email Domain Usage Analysis

This code analyzes a list of email IDs to count the usage of each email domain and calculate its percentage. It extracts the domain from each email, tallies the occurrences, and then computes the percentage of total emails for each domain.

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

### Problem Statement 5: Sales Spike Detection

This code detects significant spikes in daily sales data. It first calculates the average daily sales. Then, it identifies and displays the day number (1-indexed) and sales value for any day where sales are more than 30% above this average.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30  # 30% above average

print("Sales Spikes Detected:")
for i, sale_value in enumerate(sales):
    if sale_value > threshold:
        print(f"Day {i + 1}: {sale_value}")
```

### Problem Statement 6: Duplicate User ID Detection

This code identifies and counts duplicate user IDs from a given list. It uses a dictionary to store the frequency of each user ID and then prints only those IDs that appear more than once, along with their respective counts.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_counts = {}
for user_id in user_ids:
    id_counts[user_id] = id_counts.get(user_id, 0) + 1

print("Duplicate User IDs:")
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} \u2192 {count} times")
```
