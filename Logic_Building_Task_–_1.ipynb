## Pass / Fail Analyzer

**Problem Statement:**
Analyze student results.

**Requirements:**
*   A student passes if marks â‰¥ 50
*   Count the total number of pass students
*   Count the total number of fail students
*   Print the final result clearly
*   Real-World Application: Academic evaluation systems

```python
# Given list of marks
marks = [45, 78, 90, 33, 60, 50, 20, 85]

# Initialize counters
pass_count = 0
fail_count = 0

# Iterate through the marks and apply the passing criteria
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final result
print(f"Total students: {len(marks)}")
print(f"Number of students who Passed: {pass_count}")
print(f"Number of students who Failed: {fail_count}")
```

## Simple Data Cleaner

**Problem Statement:**
Clean and standardize user names.

**Requirements:**
*   Remove extra spaces
*   Convert all names to lowercase
*   Print the cleaned list
*   Real-World Application: Data preprocessing before analysis

```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE ", "  David ", "EVE  "]

# Create an empty list to store cleaned names
cleaned_names = []

# USing for loops to loop or iterate the names, clean them, and add to the new list
for name in names:
    # Remove extra spaces using .strip() and convert to lowercase using .lower()
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Original names:", names)
print("Cleaned names:", cleaned_names)
```

## 4. Message Length Analyzer

**Problem Statement:**
Analyze message sizes.

**Requirements:**
*   Print the length of each message
*   Flag messages longer than 10 characters
*   Real-World Application: Text filtering and validation systems

```python
# Given list of messages
messages = ["Hi", "Welcome to the platform", "OK", "Hello World!", "Short"]

# Iterate through each message
print("Message Analysis:")
for message in messages:
    message_length = len(message)
    print(f"- Message: '{message}' (Length: {message_length} characters)", end="")

    # Flag messages longer than 10 characters
    if message_length > 10:
        print(" - FLAG: Longer than 10 characters!")
    else:
        print()
```

## Logic Building Task 5: Error Message Detector

**Problem Statement:**
Detect error messages from system logs.

**Requirements:**
*   Count the number of "ERROR" entries
*   Print the total error count
*   Real-World Application: Monitoring and log analysis systems

```python
# Given list of log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR", "DEBUG", "ERROR", "INFO"]

# Initialize counters
error_count = 0
info_count = 0 # Initialize info_count here

# Iterate through the logs and count "ERROR" entries
for entry in logs:
    if entry == "ERROR":
        error_count += 1
for entry in logs:
    if entry == "INFO":
        info_count += 1

# Print the total counts
print("System Logs:", logs)
print(f"Total number of 'ERROR' entries detected: {error_count}")
print(f"Total number of 'INFO' entries detected: {info_count}")
```
