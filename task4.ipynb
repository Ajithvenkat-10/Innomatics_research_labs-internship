### Problem 1: Mobile Recharge Validation System

**Problem Statement**
Create a function to validate a mobile recharge request.

**Rules:**
* Recharge amount must be ≥ ₹50
* Valid plans: 199, 299, 399, 599
* User should be allowed to retry if invalid

**Real-Time Use**
* Telecom recharge apps
* Payment gateways

**Hint**
* Use function for validation
* Use while loop for retry

```python
def validate_recharge():
    valid_plans = {199, 299, 399, 599}

    while True:
        try:
            amount = int(input("Enter recharge amount (₹): "))
            plan = int(input("Enter plan (199, 299, 399, 599): "))

            if amount >= 50 and plan in valid_plans:
                print(f"Recharge of ₹{amount} with plan {plan} is successful!")
                break
            else:
                if amount < 50:
                    print("Recharge amount must be at least ₹50. Please retry.")
                if plan not in valid_plans:
                    print(f"Invalid plan: {plan}. Valid plans are {', '.join(map(str, valid_plans))}. Please retry.")
        except ValueError:
            print("Invalid input. Please enter a valid number for amount and plan.")

# Call the function to test
validate_recharge()
```

### Problem 2: Inventory Reorder Alert System

**Problem Statement**
Given a dictionary of products and stock quantities:
* Print 'Reorder Alert' if stock < 15
* Print 'Stock OK' otherwise

**Real-Time Use**
* Warehouse systems
* Retail management

**Hint**
* Use function
* Use for loop on dictionary

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Example usage:
products_stock = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 15,
    "Monitor": 5,
    "Webcam": 25
}

check_inventory(products_stock)
```

### Problem 3: Student Result Processing System

**Problem Statement**
Create a function that:
* Accepts a list of student marks
* Calculates average marks
* Prints 'Pass' if average ≥ 50 else 'Fail'

**Real-Time Use**
* Online exam systems
* Evaluation portals

**Hint**
* Use loop to calculate total
* Return result from function

```python
def process_student_result(marks):
    if not marks:
        print("No marks provided.")
        return

    total_marks = 0
    for mark in marks:
        total_marks += mark

    average_marks = total_marks / len(marks)

    print(f"Average Marks: {average_marks:.2f}")
    if average_marks >= 50:
        print("Result: Pass")
        return "Pass"
    else:
        print("Result: Fail")
        return "Fail"

# Example usage:
student_marks1 = [75, 80, 65, 90, 70]
process_student_result(student_marks1)

print("\n---")

student_marks2 = [40, 35, 45, 50, 20]
process_student_result(student_marks2)
```

### Problem 4: Cab Fare Estimator with Retry Option

**Problem Statement**
Create a cab fare calculator:
* Base fare = ₹50
* ₹12 per km
* Peak hour → 25% extra
* Allow retry using while loop

**Real-Time Use**
* Ride-hailing apps
* Transport systems

**Hint**
* Function for fare calculation
* While loop for retry

```python
def calculate_cab_fare(distance_km, is_peak_hour):
    base_fare = 50
    fare_per_km = 12
    peak_hour_surcharge = 0.25

    fare = base_fare + (distance_km * fare_per_km)

    if is_peak_hour:
        fare += fare * peak_hour_surcharge

    return fare

def cab_fare_estimator():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            peak_hour_input = input("Is it peak hour? (yes/no): ").lower()
            is_peak_hour = (peak_hour_input == 'yes')

            if distance < 0:
                print("Distance cannot be negative. Please retry.")
                continue

            fare = calculate_cab_fare(distance, is_peak_hour)
            print(f"Estimated Cab Fare: ₹{fare:.2f}")
            break

        except ValueError:
            print("Invalid input. Please enter a valid number for distance.")
        except Exception as e:
            print(f"An error occurred: {e}. Please retry.")

# Call the function to test
cab_fare_estimator()
```

### Problem 5: Employee Attendance Eligibility Checker

**Problem Statement**
Create a function that:
* Accepts attendance list ("P", "A")
* Calculates attendance percentage
* Returns 'Eligible' if ≥ 75% else 'Not Eligible'

**Real-Time Use**
* HR payroll systems
* Internship tracking

**Hint**
* Use loop to count present days
* Use return statement

```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        return "No attendance recorded."

    present_days = 0
    total_days = len(attendance_list)

    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {attendance_percentage:.2f}%")
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example usage:
attendance1 = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]
print(f"Eligibility: {check_attendance_eligibility(attendance1)}")

print("\n---")

attendance2 = ["P", "A", "A", "P", "A", "P", "A", "P", "A", "A"]
print(f"Eligibility: {check_attendance_eligibility(attendance2)}")
```

### Problem 6: Password Strength Checker

**Problem Statement**
Create a function to check password strength.

**Rules:**
* Minimum length = 8
* Must contain at least one digit
* Must contain at least one special character (@#$)

**Real-Time Use**
* Login systems
* Security validation

**Hint**
* Use loop over string
* Use conditions inside function

```python
def check_password_strength(password):
    min_length = 8
    has_digit = False
    has_special_char = False
    special_chars = "@#$"

    if len(password) < min_length:
        return "Weak: Password must be at least 8 characters long."

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special_char = True

    if not has_digit:
        return "Weak: Password must contain at least one digit."

    if not has_special_char:
        return f"Weak: Password must contain at least one special character from {', '.join(special_chars)}."

    return "Strong: Password meets all criteria."

# Example usage:
print(check_password_strength("MyPass123!"))
print(check_password_strength("password"))
print(check_password_strength("MyPassword"))
print(check_password_strength("MyPass123"))
print(check_password_strength("Short1!"))
print(check_password_strength("VeryStrongPassword1@"))
```
