### Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

# User Input
light_name = input("Enter Light Name: ")
light = SmartLight(light_name)
action = input("Enter Action (ON/OFF): ").upper()

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid action. Please enter ON or OFF.")

light.display_status()
```

### Problem 2: Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_card_details(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

# User Input
emp_name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
emp_department = input("Enter Department: ")
employee = EmployeeIDCard(emp_name, emp_id, emp_department)

# Expected Output
employee.display_card_details()
```

### Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact_info(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# User Input
contact_name = input("Enter Contact Name: ")
contact_phone = input("Enter Phone Number: ")
contact = Contact(contact_name, contact_phone)

# Expected Output
contact.display_contact_info()
```

### Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def generate_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# User Input
product_name = input("Enter Product Name: ")
while True:
    try:
        product_price = float(input("Enter Price: "))
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value for the price.")
product = Product(product_name, product_price)

# Expected Output
product.generate_price_tag()
```

### Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# User Input
movie_name = input("Enter Movie Name: ")
movie_rating = float(input("Enter Rating (e.g., 4.8): "))
movie = Movie(movie_name, movie_rating)

# Expected Output
movie.display_movie_details()
```

### Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# User Input
cust_name = input("Enter Customer Name: ")
del_address = input("Enter Address: ")
delivery = Delivery(cust_name, del_address)

# Expected Output
delivery.display_delivery_details()
```
